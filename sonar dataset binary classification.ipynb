```python
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier

from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import GridSearchCV

from sklearn.preprocessing import LabelEncoder
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('sonar.csv')
X = data.iloc[:,:60]
y = data.iloc[:,60:61]

num_features = X.shape[1]
print (X[0:5])
print (y[0:5])
```

## Label Encode the Class attribute

```python
encoder = LabelEncoder()
y_encoded = encoder.fit_transform(y)
print (y_encoded[0:5])
```

```python
# Set random seed
seed = 99
```

# Run 1 - Without Any Feature Engineering

```python
g_cvscore = []
g_cvstd = []

# Function to create Sequential Deep Learning Network model
def create_model(optimizer='adam'):
    nnModel = Sequential()
    nnModel.add(Dense(60, activation='relu', input_dim=num_features))
    nnModel.add(Dense(30, activation='relu'))
    nnModel.add(Dense(1, activation='sigmoid'))
    nnModel.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
    nnModel.summary()
    return nnModel

def prepare_deep_learning_model():
    # Classifier for classification
    classifier = KerasClassifier(build_fn=create_model)

    # Using Staratified for cross validation
    skFold = StratifiedKFold(n_splits=5, shuffle=True, random_state=seed)

    # Hyper parameters for tuning
    p_epochs = [100]
    hyper_params = dict(epochs=p_epochs)
    
    return (classifier, hyper_params, skFold)
```

```python
(classifier, hyper_params, skFold) = prepare_deep_learning_model()

# Execute Grid Search using classifier, hyper params and cross validation  function
gridCV = GridSearchCV(classifier, param_grid=hyper_params, cv = skFold)
results = gridCV.fit(X,y_encoded)

g_cvscore.append(results.best_score_)
g_cvstd.append(results.cv_results_['std_test_score'][results.best_index_])

print ('Best score is %s with params %s and variance is %s'%(results.best_score_, results.best_params_, results.cv_results_['std_test_score'][results.best_index_]))
```

# Run 2 - This time with some Feature Engineering

    - We see that by scaling the input data we can get a significant improvement in the accuracy of the model

```python
from sklearn.preprocessing import StandardScaler

X_scaled = StandardScaler().fit_transform(X)

(classifier, hyper_params, skFold) = prepare_deep_learning_model()

# Execute Grid Search using classifier, hyper params and cross validation  function
gridCV = GridSearchCV(classifier, param_grid=hyper_params, cv = skFold)
results = gridCV.fit(X_scaled,y_encoded)

g_cvscore.append(results.best_score_)
g_cvstd.append(results.cv_results_['std_test_score'][results.best_index_])

print ('Best score is %s with params %s and variance is %s'%(results.best_score_, results.best_params_, results.cv_results_['std_test_score'][results.best_index_]))
```

# Run 3 - Changes in Network Toplogy of Network

    - This is an illustrative change to get an intuition that changes in network topology of a neural network has significant impact on it's performance. In this example we drop down number of neurons in input layer from 60 to 30 - this negatively impacts the accuracy of the model.

```python
# Function to create Sequential Deep Learning Network model
def create_smaller_model(optimizer='adam'):
    nnModel = Sequential()
    nnModel.add(Dense(30, activation='relu', input_dim=num_features))
    nnModel.add(Dense(15, activation='relu'))
    nnModel.add(Dense(1, activation='sigmoid'))
    nnModel.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
    nnModel.summary()
    return nnModel

def prepare_small_deep_learning_model():
    # Classifier for classification
    classifier = KerasClassifier(build_fn=create_smaller_model)

    # Using Staratified for cross validation
    skFold = StratifiedKFold(n_splits=5, shuffle=True, random_state=seed)

    # Hyper parameters for tuning
    p_epochs = [100]
    hyper_params = dict(epochs=p_epochs)
    
    return (classifier, hyper_params, skFold)

(classifier, hyper_params, skFold) = prepare_small_deep_learning_model()

# Execute Grid Search using classifier, hyper params and cross validation  function
gridCV = GridSearchCV(classifier, param_grid=hyper_params, cv = skFold)
results = gridCV.fit(X_scaled,y_encoded)

g_cvscore.append(results.best_score_)
g_cvstd.append(results.cv_results_['std_test_score'][results.best_index_])

print ('Best score is %s with params %s and variance is %s'%(results.best_score_, results.best_params_, results.cv_results_['std_test_score'][results.best_index_]))
```
